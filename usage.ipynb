```python
from StorageCalculator import *
```

```python
input_file = pd.read_excel(r'C:\Users\chubin-aa\Desktop\Anaconda\StorageCalc\storage_input.xlsx')
```

```python
file = pd.read_excel('data.xlsx')
```

```python
p = []
v = []

for i in range(len(file)):
    area = round(file.loc[i, 'Площадь_100%'] * 0.85)
    h = int(round(file.loc[i, 'высота_потолков']))
    calc = StorageCalculator(input_file=input_file, s=area, height=h, max_levels=5)
    places, volumes = calc.short_run()
    p.append(places)
    v.append(volumes)
    
file['Эталонное колличество палетомест шт.'] = p
file['Эталонное количество кубометров паллетного хранения'] = v
```
